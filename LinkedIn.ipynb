```python
user_ids = df_linkedin["user_id"].tolist()
skills = df_linkedin["skills"].tolist()
```

```python
skills_temp = []
for test in skills:

    test = test.replace("[","")
    test = test.replace("]","")
    test = test.replace("'","")
    test = test.split(', ')
    skills_temp.append(test)
    
skills = skills_temp.copy()
```

```python
skill_dict = corpora.Dictionary(skills) 

# Filter out words that occur less than 20 documents, or more than 50% of the documents.
no_below = 2 # integer
no_above = 20


skill_dict.filter_extremes(no_below=no_below, no_above = no_above)

# Term Document Frequency
skill_mat = [skill_dict.doc2bow(skill) for skill in skills]

# Check the number of tokens and documents
print('Number of unique tokens: %d' % len(skill_dict))
print('Number of documents: %d' % len(skill_mat))

# skills - data_lamma
# skill_dict - id2word_1
# skill_mat - corpus_1
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3, chunksize = 50000, passes = 1):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = gensim.models.LdaMulticore(
                                          
                                           corpus = skill_mat,
                                           id2word = skill_dict,
                                           num_topics = num_topics, 
                                           random_state = 100,
                                           #update_every=1,
                                           chunksize = 50000,
                                           passes = 1,
                                           #iterations = iterations,
                                           #alpha = 'auto',
                                           #eta = 'auto',
                                           per_word_topics = True,
                                           workers  = mp.cpu_count() - 3)
                                          
                                                                                 
                                          
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=skills, dictionary=skill_dict, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python

# skills - data_lamma
# skill_dict - id2word_1
# skill_mat - corpus_1

start = 4
limit = 11
step = 2

coherence = []
for k in range(start,limit,step):
    print('Round: '+str(k))   
    
    Lda = gensim.models.LdaMulticore # gensim.models.ldamodel.LdaModel 
    ldamodel = Lda(skill_mat, 
                   num_topics=k, 
                   id2word = skill_dict, 
                   passes=5,
                   #iterations=5,  
                   chunksize = 100, 
                   eval_every = None,
                   random_state = 210224,
                   workers  = mp.cpu_count()-2)
    
    
    cm = gensim.models.coherencemodel.CoherenceModel(model=ldamodel, texts=skills,\
                                                     dictionary=skill_dict, coherence='c_v')
    coherence.append((k,cm.get_coherence()))
    
    time.sleep(5)
```

```python
import matplotlib.pyplot as plt

# Plotting the coherence values

x_val = [x[0] for x in coherence]
y_val = [x[1] for x in coherence]

plt.plot(x_val,y_val)
plt.scatter(x_val,y_val)
plt.title('Number of Topics vs. Coherence')
plt.xlabel('Number of Topics')
plt.ylabel('Coherence')
plt.xticks(x_val)
plt.show()
```
